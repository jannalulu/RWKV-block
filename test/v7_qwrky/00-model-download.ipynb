```python
# Model filename to save to
MODEL_FILENAME="qwrky7-7B.pth"

!mkdir -p "./.model"
# !wget -nc "https://huggingface.co/BlinkDL/rwkv-7-world/resolve/0629eb5928f9e929b27d1e95dd57d124d249ce6b/RWKV-x070-World-1.5B-v3-20250127-ctx4096.pth" -O "./.model/{MODEL_FILENAME}"
```

```python
# huggingface token to use
HF_TOKEN="REPLACE_TOKEN_HERE"

# This is temporary, until a public model is released
from transformers import AutoTokenizer, AutoModelForCausalLM
model = AutoModelForCausalLM.from_pretrained("recursal/QRWKV7-7B-Instruct-Test-250205" , token=HF_TOKEN, trust_remote_code=True)
```

```python
# Save the model state dict
import torch
state_dict = model.state_dict()

# Convert all to bfloat16
for k, v in state_dict.items():
    state_dict[k] = v.bfloat16()
torch.save(state_dict, f"./.model/{MODEL_FILENAME}")

# Model saved
!ls -l "./.model"
```

```python
# Log all the weights shape and dtype
for k, v in state_dict.items():
    print(f"{k}: {v.shape} {v.dtype}")
```
